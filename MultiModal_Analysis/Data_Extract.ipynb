## 왼쪽과 오른쪽 한쪽 면만 사용해서 body lean을 뽑아보자

```python
import numpy as np
    
# left shoulder x,y,z || left hip x, y, z
def projection_angle_1(x, y, z, x1, y1, z1):
    
    dx = x - x1
    dy = y - y1

    angle_radians = np.arctan2(dy, dx) 
    angle_degrees = np.degrees(angle_radians) 

    # 오른쪽에 있는 사람의 경우에는 다음과 같이 진행.
    if ((dx > 0) & (dy < 0)).all():
        angle_degrees +=180

    # 왼쪽에 있는 사람의 경우에는 180도를 더해줌. 
    if ((dx < 0) & (dy > 0)).all():
        angle_degrees +=180
        
    return angle_degrees
```

## Projection하여서 angle값을 계산해볼 것. 

```python
import numpy as np
import math

# 기울기 값을 계산. 
def slope(x1, y1, x2, y2):
    return (y2 - y1) / (x2 - x1)
    
def projection_angle(x, y, z, x1, y1, z1):
    
    #m = slope(x,y, x1, y1)
    #abs_m = abs(m)

    dx = x - x1
    dy = y - y1

    angle_radians = np.arctan2(dy, dx) 
    angle_degrees = np.degrees(angle_radians) 

    if ((dx < 0) & (dy > 0)).all():
        angle_degrees +=180

    return angle_degrees
```

##Body 값 threshold = 7값으로 설정. 

```python
# 파일 리스트 불러오기 

file = "/Users/nabi/Desktop/results/face/filtered_face_landmark/A/1W_A1/"
file_path = os.listdir(file)
out_path = "/Users/nabi/Desktop/results/face/reshape/A/1W_A1/"

if not os.path.exists(out_path):
    os.makedirs(out_path)

seq_list_file = natsort.natsorted(file_path)

if '.DS_Store' in seq_list_file:
    seq_list_file.remove('.DS_Store')

for i in seq_list_file:
    face_file = os.path.join(file, i)
    face = pd.read_excel(face_file)

    # 이상치 값 탐지
    z_scores = (face['box.height'] - face['box.height'].mean()) / face['box.height'].std()
    
    # threshold값 설정.
    threshold = 10

    face['box.height'] = np.where(np.abs(z_scores) > threshold, face['box.height'].median(), face['box.height'])

    modified_data = pd.DataFrame({'box.height' : face['box.height']})

    file_name = f"face_{i.split('.')[0]}.xlsx"
    modified_data.to_excel(os.path.join(out_path, file_name), index=False)
```

```python
step_list = ['1W', '2W', '3W', '4W'] # '2W', '3W', '4W'
human_list = ['A1', 'A2', 'A3','A4', 'B1', 'B2', 'B3', 'B4',
              'C1', 'C2', 'C3', 'C4', 'D1', 'D2', 'D3', 'D4',
              'E1', 'E2', 'E3', 'E4', 'F1', 'F2', 'F3', 'F4',
              'G1', 'G2', 'G3', 'G4']

 #'A1', 'A2', 'A3', , 'B1', 'B2', 'B3', 'B4',
            #   'C1', 'C2', 'C3', 'C4', 'D1', 'D2', 'D3', 'D4',
            #   'E1', 'E2', 'E3', 'E4', 'F1', 'F2', 'F3', 'F4',
            #   'G1', 'G2', 'G3', 'G4'

Group_list = ['A','B','C','D','E','F','G']
# ,'B','C','D','E','F','G'

# 이미지를 저장하기 위한 코드.
default_path = '/Users/nabi/Desktop/correlation/correlation_result/'

if not os.path.exists(default_path):
    os.makedirs(default_path)

#몸 기울기 값을 뽑아낼 수 있는 경로. 
body_path = "/Users/nabi/Desktop/result_lean/"
body_file_list = os.listdir(body_path)

# body데이터 순서대로 나올 수 있도록 정리.
seq_body_list = natsort.natsorted(body_file_list)

# Mac os에서 자동으로 생성되는 DS_Store 제거 메서드 
if '.DS_Store' in seq_body_list:
    seq_body_list.remove('.DS_Store')

# face 얼굴 값을 뽑아낼 수 있는 경로. 
face_path = "/Users/nabi/Desktop/results/face/reshape/"
face_list = os.listdir(face_path)

seq_face_file = natsort.natsorted(face_list)

if '.DS_Store' in seq_face_file:
    seq_face_file.remove('.DS_Store')


# 얼굴 값을 뽑아낼 수 있는 경로.
for group in Group_list:
    for step in step_list:
        # startswith() '앞에 내가 시작하기를 원하는 문자'가 있다면 그것과 같이 실행하기 위함. 
        filter_list = [human for human in human_list if human.startswith(group)]

        for human in filter_list:
            # 파일 경로 불러오기. 
            p = os.path.join(body_path, f"{group}/{step}_{human}")
            p1 = os.listdir(p)
            seq_body_list = natsort.natsorted(p1)

            if '.DS_Store' in seq_body_list:
                seq_body_list.remove('.DS_Store')

            p_ = os.path.join(face_path, f"{group}/{step}_{human}")
            p_1 = os.listdir(p_)
            seq_face_file = natsort.natsorted(p_1)

            if '.DS_Store' in seq_face_file:
                seq_face_file.remove('.DS_Store')

            for i, (body_data, face_data) in enumerate(zip(seq_body_list, seq_face_file)):
                
                file_path_ = os.path.join(p, body_data)
                body_file = pd.read_excel(file_path_)
                
                file_path_1 = os.path.join(p_, face_data)
                face_file = pd.read_excel(file_path_1)

                # X의 값은 몸의 기울기 값 할당. 
                # Y의 값은 얼굴의 박스 크기 값 할당.
                X = body_file['body_lean']
                Y = face_file['box.height']

                # 출력하고 싶은 길이 제한. 
                X = X.head(3000)
                Y = Y.head(3000)   
                
                fig, ax1 = plt.subplots()
                color_1 = 'tab:blue'
                ax1.set_title('Correlation between Body Lean and Face Height')
                ax1.set_xlabel('Index Length')
                ax1.set_ylabel('Frame(body lean)', fontsize=12, color=color_1)
                ax1.plot(X, color=color_1)
                ax1.tick_params(axis='y', labelcolor=color_1)

                ax2 = ax1.twinx() # instantiate a second axes that shares the same x-axis
                color_2 = 'tab:red'
                ax2.set_ylabel('Frame(box)', fontsize=12, color=color_2)
                ax2.plot(Y, color=color_2)
                ax2.tick_params(axis='y', labelcolor=color_2)

                fig.tight_layout()

                # # 표 위에 coefficient 이름 표시.
                # plt.legend()

                # Calculate correlation coefficient
                correlation_coefficient = X.corr(Y, method='pearson')
                correlation_text = f'Correlation Coefficient: {correlation_coefficient : .2f}'
                plt.text(0.5, 0.1, correlation_text, horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)

                #상관관계 그래프 출력하기
                # plt.show()

                #그래프 저장하는 코드 
                file_name = f'{group}_{step}_{human}_{i}.png'
                plt.savefig(os.path.join(default_path, file_name), format='png')

                plt.close()
                
```

```python
import math

def slope(x1, y1, x2, y2):
    return (y2 - y1) / (x2 - x1)

def angle_between_points(x1, y1, x2, y2):
    m = slope(x1, y1, x2, y2)
    angle_radians = math.atan(m)
    angle_degrees = math.degrees(angle_radians)
    return angle_degrees

# Given points
x1, y1 = 3, 5
x2, y2 = 7, 2

# Calculate the angle
angle = angle_between_points(x1, y1, x2, y2)

print("Angle between the line and the ground (in degrees):", angle)
```
